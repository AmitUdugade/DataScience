```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df = pd.read_csv('Downloads/Datasets - ML/canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('per_capita_income_(US$)')
plt.ylabel('year')
plt.scatter(df['per_capita_income_(US$)'], df['year'], color = 'red', marker = '+')
```

```python
model = linear_model.LinearRegression()
model
```

```python
new_df = df.drop('per_capita_income_(US$)', axis = 'columns')
new_df
```

```python
model.fit(new_df, df['per_capita_income_(US$)'])
```

```python
model.predict([[2020]])
```
