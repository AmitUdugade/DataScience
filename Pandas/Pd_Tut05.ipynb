```python
# Combine two dataframes into one excel sheet having two different sheets

df_stocks = pd.DataFrame({
    'tickers' : ['Google', 'WMT', 'Meta', 'Visa'],
    'price' : [100, 80, 70, 60],
    'eps' : [27.82, 25, 21.52, 16.02]
})

df_whether = pd.DataFrame({
    'day' : ['01/02/2018', '01/03/2019', '01/04/2020'],
    'temperature' : [32, 25, 26],
    'event' : ["Sunny", "Rain", "Cloudy"]
})
```

```python
with pd.ExcelWriter('Stocks_Whether.xlsx') as writer:
  df_stocks.to_excel(writer, sheet_name = "Stocks")
  df_whether.to_excel(writer, sheet_name = "Whether")
```
