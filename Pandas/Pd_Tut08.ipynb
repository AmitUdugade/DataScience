```python
import pandas as pd

df = pd.read_csv("/newwhether_data.csv")
df
```

```python
# Group by Cities -

g = df.groupby('city')
g
```

```python
for city, city_df in g:
  print(city)
  print(city_df)
```

```python
for city, city_dafaframe in g:
  print(city)
  print(city_dafaframe)
```

```python
#Fetch Specific city

g.get_group('Mumbai')
```

```python
#Get maximum temperature of each city

g.max()
```

```python
#Get average temperature of each city

g.mean()
```

```python
g.describe().T
```
