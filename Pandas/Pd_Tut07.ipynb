```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/whether_data.csv")
df
```

```python
#Replace -99999 into NaN by using numpy

new_df = df.replace(-99999, np.NaN)
new_df
```

```python
# if multiple value to be replace then

new_df = df.replace([-99999, -8888], np.NaN)
new_df
```

```python
new_df = df.replace({
    'temperature' : -99999,
    'windspeed' : -8888,
    'event' : '0'
}, np.NaN)

new_df
```

```python
#replace by using it's value

new_df = df.replace({
    -99999 : np.NaN,
    'no event' : 'sunny'
})

new_df
```

```python
df = pd.read_csv("/content/newwhether_data.csv")
df
```

```python
# Replace the 32F with only 32 in temperature column but don't touch in event column using REGEX

new_df = df.replace({
    'temperature' : '[A-Za-z]',
    'windspeed' : '[A-Za-z]'
}, '' , regex = True)

new_df
```
