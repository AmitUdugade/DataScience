```python
# Pivot - Pivot allows you to tranform or reshape data.

import pandas as pd

df = pd.read_csv("/newwhether_data.csv")
df
```

```python
df.pivot(index="day", columns="city")
```

```python
df.pivot(index="day", columns="city", values = "event")   #pick specific column
```

```python
# Pivot table is used to summerize and aggregate data inside dataframe.

df.pivot_table(index = 'city', columns = 'day')
```

```python
# Use aggregate function

df.pivot_table(index = 'city', columns = 'day', aggfunc = "sum")
```
