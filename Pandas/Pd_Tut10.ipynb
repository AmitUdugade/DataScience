```python
#Concat -

import pandas as pd

india_whether = pd.DataFrame({
    'city' : ["Mumbai", "Delhi", "Banglore"],
    'Temperature' : [32, 30, 31],
    'Humidity' : [80, 74, 78]
})

india_whether
```

```python
us_whether = pd.DataFrame({
    'city' : ["New York", "Chicago", "Orlando"],
    'Temperature' : [22, 25, 24],
    'Humidity' : [40, 44, 38]
})

us_whether
```

```python
df = pd.concat([india_whether, us_whether])
df
```

```python
df = pd.concat([india_whether, us_whether], ignore_index = True)
df
```

```python
# keys -

df = pd.concat([india_whether, us_whether], keys = ["India", "US"])
df
```

```python
df.loc["India"]
```

```python
temperature_df = pd.DataFrame({
    'city' : ["Mumbai", "Pune", "Nagpur"],
    'temperature' : [32, 30, 34]
})

temperature_df
```

```python
windspeed_df = pd.DataFrame({
    'city' : ["Mumbai", "Pune", "Nagpur"],
    'windspeed' : [7, 8, 5]
})

windspeed_df
```

```python
df2 = pd.concat([temperature_df, windspeed_df])
df2
```

```python
df3 = pd.concat([temperature_df, windspeed_df], axis = 1)
df3
```

```python
# make a series

s = pd.Series(["Humid","Cold","Rain"], name = 'event')
s
```

```python
#Concat the dataframe with the series -

df = pd.concat([temperature_df,  s] , axis = 1)
df
```
