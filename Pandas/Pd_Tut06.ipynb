```python
# Convert Date Column into timestamp datatype by using parse

df = pd.read_csv("/content/short_weatherHistory.csv", parse_dates = ["Date"])
type(df.Date[0])
```

```python
df.set_index('Date', inplace = True)
```

```python
df
```

```python
# Fill n.a

new_df = df.fillna(0)
new_df
```

```python
# Fill Event with 'No Event' wherever the event has 0 value.

new_df = df.fillna({
    'Event' : "No Event",
    'Temperature' : 0,
    'Wind Speed' : 0,
    'Humidity' : 0
})

new_df
```

```python
# Forward Fill

new_df1 = df.fillna(method = 'ffill')
new_df1
```

```python
# Backward Fill

new_df2 = df.fillna(method = 'bfill')
new_df2
```

```python
# Limit - carryforward the value upto 1 row

new_df = df.fillna(method = 'ffill', limit = 1)
new_df
```

```python
# Interpolate - by default method is linear

new_df = df.interpolate()
new_df
```

```python
new_df = df.interpolate(method = "time")
new_df
```

```python
# Dropna - Drop the entire row if any column has missing value.

new_df1 = df.dropna()
new_df1
```
