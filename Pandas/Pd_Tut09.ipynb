```python
# Merge dataframes

import pandas as pd

df1 = pd.DataFrame({
    'city' : ["Mumbai", "Pune", "Nagpur"],
    'Temperature' : [32, 28, 37]
})
df1
```

```python
df2 = pd.DataFrame({
    'city' : ["Nagpur", "Pune", "Mumbai"],
    'Humidity' : [65, 51, 79]
})

df2
```

```python
df3 = pd.merge(df1, df2, on = 'city')     #if both dataframes having different cities then 'ON' condition pick only common cities only. (By default - Inner join)
df3
```

```python
# Fetch common + uncommon records. use outer join.

df4 = pd.merge(df1, df2, on = 'city', how = 'outer')
df4
```

```python
# Left join.

df5 = pd.merge(df1, df2, on = 'city', how = 'left')
df5
```

```python
# Right join.

df6 = pd.merge(df1, df2, on = 'city', how = 'right', indicator = True)    # Indicator will create new column and will tell you from which table the data is fetched.
df6
```

```python
#Suffixes -

df6 = pd.merge(df1, df2, on = 'city', suffixes = ('_left','_right'))
```
