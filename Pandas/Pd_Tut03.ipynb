```python
# Different ways of creating a dataframe.

# WAY 1 - Using CSV

import pandas as pd

df = pd.read_csv("/content/whether.csv")
df
```

```python
# WAY 2 - Using Excel

df1 = pd.read_excel("/content/whether.xlsx","sheet1")
df1
```

```python
# WAY 3 - Using Python Dictionary

import pandas as pd
whether_data = {
                'Day' : ['01/01/2018','01/02/2019','01/03/2020','01/04/2021'],
                'Temperature' : [32, 35, 28, 32],
                'Windspeed' : [10, 12, 14, 16],
                'Event' : ['Rain','Sunny', 'Rain', 'Snow']
}

df2 = pd.DataFrame(whether_data)
df2
```

```python
# WAY 4 - From Tuples List

whether_data1 = [
                 ('01/01/2018', 32, 9, 'Sunny'),
                 ('01/01/2019', 20, 10, 'Snow'),
                 ('01/01/2020', 27, 9, 'Rain'),
                 ('01/01/2021', 22, 11, 'Cloudy'),
                 ]

df3 = pd.DataFrame(whether_data1, columns = ["Day","Temperature","Windspeed","Event"])
df3
```

```python
# WAY 5 - Using Dictionary List

whether_data2 = [
               {'Day' : '01/01/2018', 'Temperature' : 32, 'Windspeed' : 9, "Event" : 'Sunny'},
               {'Day' : '01/01/2019', 'Temperature' : 31, 'Windspeed' : 10, "Event" : 'Snow'},
               {'Day' : '01/01/2020', 'Temperature' : 30, 'Windspeed' : 11, "Event" : 'Cloudy'},
               {'Day' : '01/01/2021', 'Temperature' : 31, 'Windspeed' : 12, "Event" : 'Rain'},
]

df4 = pd.DataFrame(whether_data2)
df4
```
