```python
import pandas as pd

df = pd.read_csv('/content/stock_data.csv', skiprows = 1)  #it will skip row 1
df
```

```python
df1 = pd.read_csv('/content/stock_data.csv', header = 1)  #it will skip row 1 but way 2
df1
```

```python
# If you don't want Column Header

df = pd.read_csv('/content/stock_data.csv', header = None)
df
```

```python
#If you have headerless columns and want to give each column a name. or want to give different column names.
df = pd.read_csv('/content/stock_data.csv', header = None, names = ["tickers", "eps","revenue","price","people"])
df
```

```python
#if you want to read only 4 rows from the large csv file.

df = pd.read_csv('/content/stock_data.csv', skiprows = 1, nrows = 4)   #nrows - no. of lines to display without header.
df
```

```python
#Fill the n.a. values

df = pd.read_csv('/content/stock_data.csv', na_values = ["Not Available","n.a"])
df
```

```python
# Revenue can't be -1. it can be at least 0

df = pd.read_csv('/content/stock_data.csv', skiprows = 1, na_values = {
                'revenue':["Not Available","n.a", -1],
                'eps':["Not Available","n.a"],
                'price':["Not Available","n.a"],
                'people':["Not Available","n.a"],
                })
df
```

```python
# To write a CSV file.
# Following data is into df variable and we're going to write this data
# in new file called 'new_stock_data.csv'
# It will create a new csv having those data(df holding data).

df.to_csv('new_stock_data.csv')  #It will create new file but having 1 extra column called index
```

```python
#It will create new file but having 1 extra column called index so if you want to remove index column.

df.to_csv('new_stock_data.csv', index = False)
```

```python
# Import only perticular columns into new .csv file.
# It will import tickers and eps columns in a new file.
df.to_csv('selected_stock_data.csv', columns = ['tickers','eps'])
```

```python
#if you don't want headers

df.to_csv('selected_stock_data.csv', header = False, index = False)
```
